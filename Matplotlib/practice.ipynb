```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.array([1,2,3])
y=np.array([2,4,6])
plt.plot(x,y,label='2x',color='magenta',linewidth=3)

plt.title("My first graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
plt.savefig('first_graph.png',dpi=100)
```

```python
x2=np.arange(0,6,0.5)
plt.plot(x2[:4],x2[:4]**2,'r')
plt.plot(x2[3:],x2[3:]**2,'r--')
         
```

```python
#BAR CHART
labels=['a','b','c']
values=[1,5,6]
bars=plt.bar(labels,values)
plt.figure(figsize=(6,4),dpi=200)
bars[2].set_hatch('*')
plt.show()
```
