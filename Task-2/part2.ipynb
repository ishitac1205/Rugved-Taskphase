```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
f1=pd.read_csv('matches.csv')
f2=pd.read_csv('deliveries.csv')
```

```python
#18
t18=f1.groupby(['season']).count()
print(t18)
```

```python
#19
s=f2['dismissal_kind'].value_counts()
labels=s.index
values=s.values

plt.scatter(labels, values)
plt.plot(labels,values)
```

```python
#20
t20=f2.groupby(['batsman']).sum()['total_runs']
x1=t20.index[0:10]
y1=t20.values[0:10]
plt.scatter(x1,y1)
plt.plot(x1,y1)
```

```python
#21
t21=f1['player_of_match'].value_counts()
x2=t21.index[0:10]
y2=t21.values[0:10]
plt.scatter(x2,y2)
plt.plot(x2,y2)
```

```python
#22
tot=pd.concat([f1['team1'],f1['team2']])
tot_matches= tot.value_counts()
print(tot_matches)
```

```python
#24
win_matches=f1['winner'].value_counts()
x3=win_matches.index
y3=win_matches.values
plt.scatter(x3,y3)
plt.plot(x3,y3)
plt.show()
```

```python
#25

ratio=(tot_matches/win_matches)
print(ratio)

```
