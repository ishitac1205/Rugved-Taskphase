```python
import tensorflow as tf
import numpy as np

dataset_path = "/Users/ishitachoudhary/Desktop/Rugved/to show/task7/anime_characters/"
dataset = tf.keras.preprocessing.image_dataset_from_directory(dataset_path, batch_size=1)

image_sizes = set()

for images, _ in dataset:
    size = images.shape[1:3]  # Extract (height, width)
    image_sizes.add(tuple(size))

# Print unique image sizes
print("Unique image sizes:", image_sizes)

if len(image_sizes) == 1:
    print("✅ All images have the same size.")
else:
    print("❌ Images have different sizes.")
```

```python
import os

dataset_path = "/Users/ishitachoudhary/Desktop/Rugved/to show/task7/anime_characters/"

# Dictionary to store the count of images per class
image_counts = {}

for subdir, _, files in os.walk(dataset_path):
    if subdir != dataset_path:  # Ignore the root folder
        class_name = os.path.basename(subdir)
        num_images = len([f for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))])
        image_counts[class_name] = num_images

# Print the results
for class_name, count in image_counts.items():
    print(f"📂 {class_name}: {count} images")
```
